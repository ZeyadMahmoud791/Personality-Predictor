```python
# import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score


# Load iris training dataset
iris = datasets.load_iris()

# feature of the dataset
X = iris.data
# target of the dataset
Y = iris.target

# split dataset into training dataset and test dataset
x_train ,x_test ,y_train , y_test = train_test_split(X,Y)

# # PLot data points
# plt.scatter(x[:,0], x[:,1], c=y)
# plt.show()
                                                            #(100,)
# Create object of type MLPClassifier to access the needed functions
neuralNetworkClassifier = MLPClassifier(hidden_layer_sizes = (100,200, 30), max_iter = 5000, solver='lbfgs', shuffle = True
                                        , learning_rate_init = 0.1)

# Solver 
# lbfgs’ is an optimizer in the family of quasi-Newton methods.
# ‘sgd’ refers to stochastic gradient descent.
# ‘adam’ refers to a stochastic gradient-based optimizer proposed by Kingma, Diederik, and Jimmy Ba

# shuffle = true : to shuffle dataset

# Pass the training dataset and its corresponding targets(classes)
neuralNetworkClassifier.fit(x_train, y_train)

# Predict the targets of the test dataset
y_pred = neuralNetworkClassifier.predict(x_test)

# Compare the true targets with the predicted targets
# by computing the accuracy
print(accuracy_score(y_test, y_pred))
```

```python
from sklearn import datasets
from sklearn.linear_model import Perceptron
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

# Load iris training dataset
iris = datasets.load_iris()

# feature of the dataset
X = iris.data
# target of the dataset
Y = iris.target

# split dataset into training dataset and test dataset
x_train ,x_test ,y_train , y_test = train_test_split(X,Y)

# Create object of type perceptronClassifier to access the needed functions
perceptronClassifier = Perceptron()

# Pass the training dataset and its corresponding targets(classes)
perceptronClassifier.fit(x_train, y_train)

# Predict the targets of the test dataset
y_pred = perceptronClassifier.predict(x_test)

# Compare the true targets with the predicted targets
# by computing the accuracy
print(accuracy_score(y_test, y_pred))
```

```python
from sklearn import datasets
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score

# Load iris training dataset
iris = datasets.load_iris()

# Create object of type GaussianNB to access the needed functions
gnb = GaussianNB()

# Pass the training dataset and its corresponding targets(classes)
gnb.fit(iris.data, iris.target)

# Predict the targets of the test dataset
y_pred = gnb.predict(iris.data)

# Compare the true targets with the predicted targets
# by computing the accuracy
print(accuracy_score(iris.target, y_pred))
```
